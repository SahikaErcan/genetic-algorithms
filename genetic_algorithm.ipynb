```python
import time
import random
```

```python
class Genetic:
    def __init__(self, Target, Population_Number):
        self.Gene_Pool = """abcçdefgğhıijklmnoöprsştuüvwxyzABCDEFGĞHIİJKLMNOÖPRSŞTUÜVWXYZ 1234567890,.-;:_!"^%&/()=?@${[]}"""
        self.Target = Target
        self.Population_Number = Population_Number
        self.Target_Text_Length = len(Target) # dışardan gelen text in uzunluğuna ihtiyacımız var.
        self.Population = [] # popülasyonu bu liste içinde tutacağız.
        self.Next_Generation = [] # çiftleşme yaparken işimize yarayacak
        self.Found = False # hedefe ulaşıp ulaşmadığımızı kontrol edeceğiz.
        self.Generation_Timer = 0 # Kaç adım ilerlediğimizi göreceğiz.
        
    class Member: # Bireylerimizi tanımlayalım.
        def __init__(self, Chromosome): # üyenin dna bilgilerini taşıyan kromozom bilgisi olacak
            self.Chromosome = Chromosome
            self.Fitness = 0 # Sağlık durumu başlangıçda 0 
                
    def random_gene(self): # Gen oluşturuyoruz.
        Gene = random.choice(self.Gene_Pool) # Gene_Pool dan 1 tane gen seç.
        return Gene
    def create_chromosome(self): # Kromozom oluşturuyoruz.
        chromosome = [self.random_gene() for i in range(self.Target_Text_Length)]
        return chromosome
    
    def calculate_fitness(self): # Popülasyondaki tüm bireylerin hedefe yakınlık-sağlık durumunu hesaplıyoruz.
        for Member in self.Population:
            Member.Fitness = 0 # Önceki nesilden gelmiş olabilir ve tekrar değerlendirmeye geldiğinde fitness ın üstüne yapacağımız eklemeler belirli bir sayıdan değil 0 dan olsun istiyoruz.
            for i in range(self.Target_Text_Length): # Kromozom ile Targeti karşılaştırabilmek için target uzunluğu kadar dönüyoruz.
                if Member.Chromosome[i] == self.Target[i]: # Her eşitlik durumunda ilgili üye için yakınlık durumu artmış olacak
                    Member.Fitness += 1
            # Üyenin fitness değeri targetın uzunluk değerine eşitse bu her elemanın aynı olduğu anlamına gelir.
            # Her eleman aynı ise hedefimize ulaşmış oluyoruz.
            if Member.Fitness == self.Target_Text_Length:
                self.Found_Text = Member.Chromosome
                self.Found = True
    
    # Sıralanmış dizinin sondaki değerleri alıp diğer nesile geçirip bunları çiftleştirerek yeni üyeler oluşturmamız gerekiyor.
    def crossover(self): # Çiftleşme Fonksiyonu
        last_best = int((90 * self.Population_Number) / 100) # son %10 unu almak için belirli bir sayı bulmamız gerekiyor.
        
        self.Next_Generation = []
        self.Next_Generation.extend(self.Population[last_best:])
        
        while True:
            if len(self.Next_Generation) < self.Population_Number:
                member_1 = random.choice(self.Population[last_best:]).Chromosome
                member_2 = random.choice(self.Population[last_best:]).Chromosome
                new_member = []
                
                for gene1, gene2 in zip(member_1, member_2):
                    prob = random.random() # 0-1
                    if prob < 0.47:
                        new_member.append(gene1)
                    elif prob < 0.94:
                        new_member.append(gene2)
                    else:
                        new_member.append(self.random_gene())
                        
                self.Next_Generation.append(self.Member(new_member))
                
            else:
                break
        
        self.Population = self.Next_Generation
    
    def main(self):
        # Yapacağımız ilk şey bir başlangıç populasyonu olşturmaktır.
        for i in range(self.Population_Number): # Boş dizimize Populasyon sayısı kadar üye ekleyeceğiz.
            self.Population.append(self.Member(self.create_chromosome())) # kromozom oluşturarak bir üye elde ediyoruz. Elde ettiğimiz üyeyi de Populasyon dizisine ekliyoruz.
        
        while not self.Found: # Found yani bulundu değişkeni True olana kadar devam edecek
            self.calculate_fitness()
            # sağlılkı olanlara ulaşabilmek için fitness değerini sıralayalım.
            self.Population = sorted(self.Population, key=lambda Member: Member.Fitness) # memberin fitness değerine göre sırala diyoruz.
            self.crossover()
            self.Generation_Timer += 1
            
            """fitness = []
            for member in self.Population: # her üyenin üzerinde dönelim.
                fitness.append(member.Fitness)
                
            print(fitness)"""
        
        print(f"You Found = {self.Found_Text}, you did it {self.Generation_Timer} steps."
              f"\nYour Target = {self.Target}")
        
        
Target = "Mustafa Kemal Atatürk"
Population_Number = 500
Go = Genetic(Target, Population_Number)
Go.main()

```
